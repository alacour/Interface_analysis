```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import freud
from scipy.stats import uniform
```

```python
# Reading File

f = open("recombine.lammpstrj")
text = f.readlines()
f.close()
```

```python
# Making Array of positions


print(len(text))
poses = []

atom_number = 3066
for i in range(0, 40000, 10): #Only using 1/10 of the frames
    it = 9 + (9+atom_number)*i # The preamble to positions is 9 lines long
    et = it + (atom_number)
    print(i,it, et, et-it)
    pos = []
    for j in range(it, et):
        pos.append(np.fromstring(text[j][:-1], sep=' '))

    poses.append(np.asarray(pos))

                
                
            #break
        
        
        
        
#box = [62.22*2, 18.666*2, 18.666*2]        
```

```python
box = [12.438*2, 12.438*2, 12.439*2]
fbox = freud.box.Box.from_box(box)

cocos = []
uncos = []
OH3locs = []
OHlocs = []
OH3pos = []
OHpos = []

modposes = []


for it, pos in enumerate(poses):
    #print(pos)
    whO = np.where(pos[:,0] == 2)[0]
    whH = np.where(pos[:,0] == 1)[0]
    whE = np.where(pos[:,0] == 3)[0]

    Opos = pos[whO, 2:5]
    Hpos = pos[whH, 2:5]


    aq = freud.locality.AABBQuery(box, Hpos)
    nlist = aq.query(Opos, {'r_max': 1.4}).toNeighborList() # Cutoff for being OH3 is having 3 neighbors in 1.4A
    un, co = np.unique(nlist.query_point_indices, return_counts=True)
    unco, coco = np.unique(co, return_counts=True)


    loc = un[np.where(co == 3)] # checking whether an oxygen has three neighbors

    modpos = np.copy(pos)

    if len(loc) > 0:
        OH3locs.append([it, loc[0]]) # If multiple oxygen have three neighbors (rare), I only call one OH3
        OH3pos.append([it, Opos[loc][0]])
        whc = [whO[loc]][0]
        for wh in whc:
            modpos[wh][0] = 5 # I assign the oxygen of the OH3 a new type

      
    nlist = aq.query(Opos, {'r_max': 1.3}).toNeighborList() # Cutoff for being OH is having 1 neighbors in 1.3A
    un, co = np.unique(nlist.query_point_indices, return_counts=True)
    unco, coco = np.unique(co, return_counts=True)
    
    
    loc = un[np.where(co == 1)] # checking whether an oxygen has one neighbor
    if len(loc) > 0:

        #print(un, loc)
        OHlocs.append([it, loc[0]]) # If multiple oxygen have one neighbor (rare), I only call one OH
        OHpos.append([it, Opos[loc][0]])
        whc = [whO[loc]][0]
        for wh in whc:
            modpos[wh][0] = 6   # I assign the oxygen of the OH a new type




    cocos.append(coco)
    uncos.append(unco)

    modposes.append(modpos)
    print(it, unco, coco) # amount of waters identified as having 1, 2, or 3 neighbors.
    # if they all have 2, the OH3 and OH have recombined

    
uncos = np.asarray(uncos, dtype='object')
cocos = np.asarray(cocos, dtype='object')
OH3pos = np.asarray(OH3pos, dtype='object')
OHpos = np.asarray(OHpos, dtype='object')

#print(len(OHpos), len(OH3pos))
```

```python
# Setting up distance calculation from OH to OH3. 
#I only use frames where both OH and OH3 are clearly identifiable

OH3frames = np.unique(OH3pos[:,0])
OHframes = np.unique(OHpos[:,0])

bothframes = np.intersect1d(OHframes, OH3frames) # frames where they are both identifiable

print(len(OHpos), len(OH3pos), len(bothframes))

ion_pos = []

for frame in bothframes:
    OH3 = OH3pos[OH3frames == frame, 1][0]
    OH = OHpos[OHframes == frame, 1][0]
    
    ion_pos.append([OH3, OH])
    
ion_pos = np.array(ion_pos)
#print(ion_pos)
ion_dis = fbox.wrap(ion_pos[:,1] - ion_pos[:,0])
ion_dis = np.linalg.norm(ion_dis, axis=1)

plt.plot(bothframes, ion_dis)

plt.xlabel("Frame #")
plt.ylabel("H$_3$O$^+$ to OH$^-$ distance")
```
